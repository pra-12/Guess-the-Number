```python
import random
import time

class GuessingGame:
    def __init__(self):
        self.name = ""

    def get_name(self):
        print("May I ask you for your name?")
        self.name = input()

    def intro(self):
        print(f"{self.name}, we are going to play a game. I am thinking of a number between 1 and 200")
        time.sleep(0.5)
        print("Go ahead. Guess!")

    def play_game(self):
        number = random.randint(1, 200)
        guesses_taken = 0

        while guesses_taken < 6:
            time.sleep(0.25)
            enter = input("Guess: ")

            try:
                guess = int(enter)

                if 1 <= guess <= 200:
                    guesses_taken += 1

                    if guesses_taken < 6:
                        if guess < number:
                            print("The guess of the number you entered is too low.")
                        elif guess > number:
                            print("The guess of the number you entered is too high.")
                        elif guess != number:
                            time.sleep(0.5)
                            print("Try Again!")

                    if guess == number:
                        break

                else:
                    print("Silly Goose! That number isn't in the range!")
                    time.sleep(0.25)
                    print("Please enter a number between 1 and 200")

            except ValueError:
                print(f"I don't think that {enter} is a number. Sorry")

        if guess == number:
            guesses_taken = str(guesses_taken)
            print(f'Good job, {self.name}! You guessed my number in {guesses_taken} guesses!')

        if guess != number:
            print(f'Nope. The number I was thinking of was {number}')

def main():
    play_again = "yes"
    while play_again.lower() in ("yes", "y"):
        game = GuessingGame()
        game.get_name()
        game.intro()
        game.play_game()
        play_again = input("Do you want to play again? (yes/no): ")

    print("Thank you for playing the guessing game!")

if __name__ == "__main__":
    main()
```
